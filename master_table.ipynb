```python
import pandas as pd
import numpy as np
```

```python
delivery_times = pd.read_csv("delivery_times_actual_duration.csv")

#convert datetime column
delivery_times['picked_up_time'] = pd.to_datetime(delivery_times['picked_up_time'])
delivery_times['delivered_time'] = pd.to_datetime(delivery_times['delivered_time'])

```

```python
users = pd.read_csv(r"C:\Users\asus\Desktop\OZI project\csv\users.csv")
orders = pd.read_csv(r"C:\Users\asus\Desktop\OZI project\csv\orders.csv")
feedback = pd.read_csv(r"C:\Users\asus\Desktop\OZI project\csv\feedback.csv")
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
```

```python
#merge order with delivery time
orders_with_delivery = orders.merge(
    delivery_times[['order_id', 'actual_delivery_mins']],
    on='order_id',
    how='left'
)

#merge user data
orders_with_feedback = orders_with_delivery.merge(
    feedback,
    on='order_id',
    how='left'
)

#merge feedback data
master_orders_df = orders_with_feedback.merge(
    users,
    on='user_id',
    how='left'
)
```

```python
master_orders_df.to_csv(
    "master_orders_dataset.csv",
    index=False
)

print("✅ master_orders_dataset.csv saved successfully")
```
