```python
import pandas as pd
import numpy as np
```

```python
master_orders_df=pd.read_csv(r"C:\Users\asus\Desktop\OZI project\master_orders_dataset.csv")
happy_high_value_users_january_df=pd.read_csv(r"C:\Users\asus\Desktop\OZI project\happy_high_value_users_january.csv")
feb_orders = master_orders_df[
    (master_orders_df['order_date'] >= '2024-02-01') &
    (master_orders_df['order_date'] <= '2024-02-28')
]

#unique feb users
feb_users = feb_orders['user_id'].unique()
```

```python
# January cohort
jan_users = set(happy_high_value_users_january_df['user_id'])

# February buyers
feb_users = set(feb_orders['user_id'])

# Retained users
retained_users = jan_users.intersection(feb_users)

retained_user_count = len(retained_users)

retention_rate = (
    retained_user_count / len(jan_users) * 100
    if len(jan_users) > 0 else 0
)

retained_user_count, retention_rate
```

```python
#count of retained users
retained_user_count = len(retained_users)
retained_user_count

#save as csv
retained_users_df = pd.DataFrame({
    'user_id': list(retained_users)
})

# Save retained users to CSV
retained_users_df.to_csv(
    "retained_users_february.csv",
    index=False
)

print("âœ… retained_users_february.csv saved successfully")
print("Retained Users (Feb):", len(retained_users))
```
