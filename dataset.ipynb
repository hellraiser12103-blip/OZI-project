```python
import pandas as pd
import numpy as np
```

```python
users = pd.read_csv(r"C:\Users\asus\Desktop\OZI project\csv\users.csv")
orders = pd.read_csv(r"C:\Users\asus\Desktop\OZI project\csv\orders.csv")
order_events = pd.read_csv(r"C:\Users\asus\Desktop\OZI project\csv\order_events.csv")
feedback = pd.read_csv(r"C:\Users\asus\Desktop\OZI project\csv\feedback.csv")
```

```python
#shape_check
print("Users:", users.shape)
print("Orders:", orders.shape)
print("Order Events:", order_events.shape)
print("Feedback:", feedback.shape)
```

```python
#Convert dates in orders
orders['order_date'] = pd.to_datetime(orders['order_date'])
```

```python
#Convert timestamps in order_events
order_events['event_timestamp'] = pd.to_datetime(order_events['event_timestamp'])
```

```python
#Check Missing Values
print("Users Missing Values:\n", users.isnull().sum())
print("\nOrders Missing Values:\n", orders.isnull().sum())
print("\nOrder Events Missing Values:\n", order_events.isnull().sum())
print("\nFeedback Missing Values:\n", feedback.isnull().sum())
```

```python
#Check Duplicate users
users.duplicated(subset=['user_id']).sum()

#Duplicate orders
orders.duplicated(subset=['order_id']).sum()

#Duplicate events (same order, same event, same timestamp)
order_events.duplicated(subset=['order_id', 'event_id', 'event_timestamp']).sum()
```

```python
#Remove Exact Duplicate Event Logs 
order_events = order_events.drop_duplicates(
    subset=['order_id', 'event_id', 'event_timestamp']
)
```
